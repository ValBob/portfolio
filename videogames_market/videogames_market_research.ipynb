# Описание проекта

**Имеются:**
- исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы  

**Нужно:**
- выявить определяющие успешность игры закономерности для создания потенциально успешного продукта и планирования рекламных акций

**Описание данных:**
- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)

- `Other_sales` — продажи в других странах (миллионы проданных копий)

**Обозначения рейтинга ESRB:**
- «`E`» («Everyone») — «Для всех» (ранее "`K-A`" ("Kids to Adults")) (Возраст `0+`)
- «`EC`» («Early childhood») — «Для детей младшего возраста» (Возраст `3+`)
- «`E10+`» («Everyone 10 and older») — «Для всех от 10 лет и старше» (Возраст `10+`)
- «`T`» («Teen») — «Подросткам» (Возраст `13+`)
- «`M`» («Mature») — «Для взрослых» (Возраст `17+`)
- «`AO`» («Adults Only 18+») — «Только для взрослых» (Возраст `18+`)
- «`RP`» («Rating Pending») — «Рейтинг ожидается»

**Предварительные замечания по рейтингу ESRB:**
- есть две категории `E` и `K-A` (устаревшее значение), которые следует объединить в одну;
- в категорию `RP` можно будет отнести все отсутствующие значения;
- можно сопоставить категориальным значениям `рейтинга` ранговые значения `минимального возраста`.

## Изучение выпуска игр по годам

Посмотрите, сколько игр выпускалось в разные годы. Важны ли данные за все периоды?

```python
def aggfunc_by_year(df=games_df, aggfunc='sum', color='blue', ylabel="Продажи, млн.копий"):
    """
    Строит столбчатую диаграмму распределения по годам
    по заданной аггрегирующей функции.
    """
    plt.figure(figsize=(10, 6))
    data=df.groupby('year_of_release').agg(sales=('total_sales', aggfunc))
    sns.barplot(data=data, x=data.index.astype('int'),
                y='sales', color=color, alpha=.5)
    plt.xticks(rotation='vertical')
    plt.ylabel(ylabel)
    plt.xlabel("По годам");
```

```python
# распределение по годам количества проданных игр (млн.копий)
aggfunc_by_year()
```

```python
# распределение по годам количества проданных игр (млн.копий)
aggfunc_by_year(aggfunc='count', color='green', ylabel="Количество релизов игр")
```

**Выводы:**

## Изучение объема продаж и времени жизни игровых платформ

Посмотрите, как менялись продажи по платформам. Выберите платформы с наибольшими суммарными продажами и постройте распределение по годам. За какой характерный срок появляются новые и исчезают старые платформы?

```python
def topn_platforms(df, top_n=None):
    """Группирует объемы продаж по платформе и
    возвращает заданное число платформ с наибольшими продажами"""
    # получим сводные данные по объемам продаж игр по платформам
    df = (df.groupby('platform')  # группируем по платформе
     .agg(total_sales=('total_sales', 'sum'))  # суммируем значения общих продаж
     .sort_values('total_sales', ascending=False)  # сортируем полученные суммы по убыванию
    )
    # добавим столбец по доле продаж по конкретной платформе к общему объёму продаж
    df['sales_share'] = df.eval('total_sales / total_sales.sum()')
    # выведем ТОП_N игр по объему продаж
    return df[:top_n]
```

```python
# применим функцию для просмотра 10 наиболее популярных платформ
topn_platforms(games_df, top_n=10)
```

```python
# вычислим долю, которую занимают ТОП10 в общем объёме продаж
topn_platforms(games_df, top_n=10).sum()
```

Доля `ТОП10 платформ` в общем объёме продаж составляет `76%`.

```python
# массив ТОП10 платформ
best_platforms = topn_platforms(games_df, top_n=10).index.values
```

```python
from matplotlib import rcParams as rcp
```

```python
# поправим умолчанния для красибого отображения меток года
rcp['xtick.direction'] = 'in'
```

```python
# построим графики распределения объемов продаж по годам для платформ из ТОП10
fig, ax = plt.subplots(nrows=5, ncols=2, figsize=(10, 15), gridspec_kw={'hspace':.35})
for platform, axi in zip(best_platforms, ax.flatten()):
    (games_df[games_df['platform'] == platform]
     .groupby('year_of_release')
     .agg(total_sales=('total_sales', 'sum'))
     .plot(kind='bar', ax=axi, title = platform, legend=False, xlabel='',
           fontsize=8))
```

**Выводы:**
- "Средним" временем жизни платформы является 8 лет.
- Из 10 наиболее популярных платформ только на 2 выпускались игры в 2016 году.
- Для прогнозироание следует взять период, начиная с 2008 года.

## Выбор потенциальных платформ за актуальный период

```python
# данные за актуальный период (с 2008 года)
games_df_actual = games_df[games_df['year_of_release'] >= 2008]
```

```python
# ТОП10 платформ за актуальный период
top10_after2008 = topn_platforms(games_df_actual, top_n=10)
top10_after2008
```

```python
# вычитанием множеств найдем платформы, ранее не попадавшие в ТОП10
new_platforms = list(set(top10_after2008.index.values) - set(best_platforms))
new_platforms
```

```python
# построим графики распределения объемов продаж по годам за актуальный период
# для новых платформ
fig, ax = plt.subplots(ncols=2, figsize=(10, 3), gridspec_kw={'hspace':.35})
for platform, axi in zip(new_platforms, ax.flatten()):
    (games_df[games_df['platform'] == platform]
     .groupby('year_of_release')
     .agg(total_sales=('total_sales', 'sum'))
     .plot(kind='bar', ax=axi, title = platform, legend=False, xlabel='',
           fontsize=8))
```

**Выводы:**
- После ограничения начала актуального периода 2008 годом в ТОП10 платформ с наибольшим объёмом продаж были включены более новые платформы `3DS`, `XOne`.
- Из всех платформ с большим объёмом продаж находятся `на максимуме` (с учетом, что за 2016 год данные представлены не полностью) две платформы: `PS4`, `XOne`.
- Продажи по `остальным` рассмотренным платформам `уменьшаются`.

# Портрет пользователя по регионам

```python
def pie_top5(by='genre', what='na_sales', how='sum', ax=None):
    """Выводит круговую диаграмму для ТОП5 категорий"""
    (games_df_actual.groupby(by)
     .agg({what: how})
     .sort_values(what, ascending=False)[:5]
     .plot.pie(subplots=True, legend=False, ylabel='', ax=ax,
               autopct='%1.1f%%', fontsize=8)
    )
    ax.set_title(what +' by '+ by);
```

```python
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(10, 10))
ax = ax.flatten()
i = 0
for row in ['platform', 'genre', 'rating']:
    for sale in ['na_sales', 'eu_sales', 'jp_sales']:
        pie_top5(by=row, what=sale, ax=ax[i])
        i += 1
```

**Выводы:**

 - Пользователи в Европе и Северной Америке имееют схожие предпочтения в использовании игровых платформ: наиболее популярные платформами являются `X360` и `PS3` (при этом в Северной Америке лидирует `X360`, а в Европе - `PS3`). В Японии лидируют платформы `3DS` и `DS` соответственно (`PS3` занимает только 3 место).
- *Предпочтения по жанрам*

В `Японии` пользователи предпочитают приключенчиский жанр (`Adventure`), который связан с построением стратегии для прохождения игры. Данные различия в предпочтениях американских и европейских пользователей от пользователей из Японии можно объяснить культурными особенностями.

# Проверка гипотез

```python
from scipy import stats
```

## Гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

**Нулевая гипотеза (Н0):** Рейтинги одинаковые  
**Альтернативная гипотеза (Н1):** Рейтинги отличаются

```python
# построим диаграмму плотности частот изучаемых распределений
x_one = games_df_actual.loc[games_df_actual['platform'] == 'XOne', 'user_score']
pc = games_df_actual.loc[games_df_actual['platform'] == 'PC', 'user_score']
ax = x_one.plot(kind='kde')
pc.plot(kind='kde', ax=ax)
plt.legend(['XOne', 'PC'])
plt.ylabel('')
plt.title('Рейтинг пользователей по платформам');
```

Распределение похоже на "нормальное", дисперссии сопоставимы применим параметрический тест Стьюдента

```python
stats.ttest_ind(x_one, pc)
```

<div class="alert alert-info">
<b>⚠️ Комментарий студента:</b> 
    <br>Вычисление средних значений выборок занимает время!
    <br>Можно пропустить 2 следующие строки, если не нужна визуализация.
</div>

```python
# построим гистограмму средних выборок для визуализации различия
x_one_means = np.fromiter((x_one.sample(n=1000, replace=True).mean()
                        for _ in range(100000)), dtype='float')
pc_means = np.fromiter((pc.sample(n=1000, replace=True).mean()
                        for _ in range(100000)), dtype='float')
```

```python
plt.figure(figsize=(10,8))
plt.hist(x_one_means, bins=50, alpha=.5)
plt.hist(pc_means, bins=50, alpha=.5)
plt.legend(['XOne', 'PC']);
```

**Вывод:**  
- Уровень значимости (p-value) значительно больше уровня (0.05).
- Нулевая `гипотеза о равенстве` средних рейтингов по платформ `XOne` и `PC` `принимается`.

## Гипотеза: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

**Нулевая гипотеза (Н0):** Рейтинги одинаковые  
**Альтернативная гипотеза (Н1):** Рейтинги отличаются

```python
# построим диаграмму плотности частот изучаемых распределений
action = games_df_actual.loc[games_df_actual['genre'] == 'Action', 'user_score']
sports = games_df_actual.loc[games_df_actual['genre'] == 'Sports', 'user_score']
ax = action.plot(kind='kde')
sports.plot(kind='kde', ax=ax)
plt.legend(['Action', 'Sports'])
plt.ylabel('')
plt.title('Рейтинг пользователей по жанрам');
```

```python
stats.ttest_ind(action, sports)
```

<div class="alert alert-info">
<b>⚠️ Комментарий студента:</b> 
    <br>Вычисление средних значений выборок занимает время!
    <br>Можно пропустить 2 следующие строки, если не нужна визуализация.
</div>

```python
# построим гистограмму средних выборок для визуализации различия
action_means = np.fromiter((action.sample(n=1000, replace=True).mean()
                        for _ in range(100000)), dtype='float')
sports_means = np.fromiter((sports.sample(n=1000, replace=True).mean()
                        for _ in range(100000)), dtype='float')
```

```python
plt.figure(figsize=(10,8))
plt.hist(action_means, bins=50, alpha=.5)
plt.hist(sports_means, bins=50, alpha=.5)
plt.legend(['action', 'sports']);
```

**Вывод:**  
- Уровень значимости (p-value) значительно меньше уровня (0.05).
- Нулевая гипотеза о равенстве средних рейтингов по жанрам Action и Sports отклоняется.
- Принимается альтернативная гипотеза, что `пользовательские рейтинги разные`.
