### Проверка правильности расчёта эффективности обогащения

- Проверьте, что эффективность обогащения рассчитана правильно.
- Вычислите её на обучающей выборке для признака `rougher.output.recovery`.
- Найдите $MAE$ между вашими расчётами и значением признака.
- Опишите выводы.

$Recovery = \frac {C * (F - T)}{F * (C - T)} * 100\%$
- Recovery - эффективность обогащения
- C - доля золота после очистки
- F - доля золота до очистки
- T - доля золота в хвостах

```python
def recovery(df: pd.DataFrame, after: str, before: str, tail: str) -> pd.Series:
    """Вычисляет эффективность обогащения"""
    after, before, tail = df[after], df[before], df[tail]
    return after * (before - tail) / (before * (after - tail)) * 100
```

```python
#Средняя абсолютная ошибка вычисленной эффективности обогащения на обучающей выборке
mean_absolute_error(train['rougher.output.recovery'],
                    recovery(df=train,
                             after='rougher.output.concentrate_au',
                             before='rougher.input.feed_au',
                             tail='rougher.output.tail_au'))
```

**Вывод:**
___
- Средняя абсолютная ошибка практически равна нулю.
- Эффективность обогащения вычислена `корректно`.

## Модель

### Функция для вычисления итоговой sMAPE.

$sMAPE = \frac 1N \sum_{i=1}^N \frac {|y_i - \hat{y_i}|}{(|y_i| + |\hat{y_i}|) / 2} * 100\%$

```python
def smape(target: pd.Series, predictions: pd.Series) -> float:
    """Вычисляет симметричную процентную оценку среднюю абсолютную ошибку"""
    return (abs(target - predictions) / (abs(target) + abs(predictions))).mean() * 50
```

```python
ytrain.values[:, 0]
```

```python
smape(ytrain.values[:, 0], ytrain.values[:, 0] + 1)
```

$sMAPE_{estimate} = 25\% * sMAPE_{rougher} + 75\% * sMAPE_{final}$

```python
def smape_estima(targets: pd.Series, predictions: np.array,
                 estima: list=[.25, .75]) -> float:
    """
    Вычисляет взвешенную sMAPE по нескольким признакам:
    targets - pd.Series целевых значений,
    predictions - np.array предсказаний,
    estima - список весов, присваиваемых оценкам.
    """
    return sum(smape(targets.iloc[:, col], predictions[:, col]) * e
               for col, e in enumerate(estima))
```

```python
# создадим оцениватель из функции smape_estima
score = make_scorer(smape_estima, greater_is_better=False)
```
